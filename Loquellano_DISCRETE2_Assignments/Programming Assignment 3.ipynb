1. Walk: A sequence of vertices where consecutive pairs are connected by edges.<br>
2. Trail: A walk where no edge is repeated.<br>
3. Path: A trail where no vertex is repeated.<br>
4. Closed Walk: A walk that starts and ends at the same vertex.<br>
5. Circuit: A closed walk where no vertex (except possibly the starting/ending vertex) is repeated.<br>
6. Simple Circuit: A circuit where no vertex or edge (except the starting/ending vertex) is repeated.<br>

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5)])

nx.draw(G, with_labels=True)

degrees = dict(G.degree())
print("Degrees of each vertex:")
for node, degree in degrees.items():
    print(f"Vertex {node}: Degree {degree}")

# Calculate the sum of the degrees of all vertices
degree_sum = sum(degrees.values())

# Calculate twice the number of edges
twice_edges = 2 * G.number_of_edges()

# Verify the Handshake Theorem
print(f"\nSum of degrees of all vertices: {degree_sum}")
print(f"Twice the number of edges: {twice_edges}")

plt.show()
```

The Handshake Theorem states that in any graph, the sum of the degrees of all vertices is equal to twice the number of edges. It's called the handshake theorem because you can imagine each edge as a handshake between two vertices, so the total number of handshakes is twice the number of edges.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a connected graph
G_connected = nx.Graph()
G_connected.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5)])

# Create another graph that is its connected components
connected_components = list(nx.connected_components(G_connected))
G_disconnected = nx.Graph()
for component in connected_components:
    G_disconnected.add_edges_from(nx.utils.pairwise(component))

# Draw the connected graph
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
pos_connected = nx.spring_layout(G_connected)  # Get layout of connected graph
nx.draw(G_connected, pos=pos_connected, with_labels=True, node_color='skyblue', node_size=1000, font_size=12, font_weight='bold')
plt.title('Connected Graph')

# Draw the disconnected graph with the same positions
plt.subplot(1, 2, 2)
nx.draw(G_disconnected, pos=pos_connected, with_labels=True, node_color='salmon', node_size=1000, font_size=12, font_weight='bold')
plt.title('Connected Components')

plt.tight_layout()
plt.show()
```

### Connected Graph: <br>
- A connected graph is a graph where there exists a path between every pair of vertices. In other words, there are no isolated vertices or groups of vertices that are not connected to the rest of the graph. <br>
### Connected Components: <br>
- Connected components are the maximal subgraphs of a graph in which every pair of vertices is connected by a path. In other words, a connected component is a subset of vertices in a graph such that there is a path between every pair of vertices in the subset, and this subset is not connected to any other vertex outside the subset.<br>


### Euler's Circuit <br>

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.cycle_graph(8)
euler_circuit = list(nx.eulerian_circuit(G))

# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1000, font_size=12, font_weight='bold')

plt.title(f"Has Eulerian Circuit: {euler_circuit}")
plt.show()
```

An Eulerian circuit is a path in a graph that traverses every edge exactly once and returns to the starting point. In simpler terms, it's a route through a graph that covers all the edges without repeating any, allowing you to start and end at the same node.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph with an Eulerian trail
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (4, 5),(1, 5)])  # Example of an Eulerian trail

# Get an Eulerian trail
euler_trail = list(nx.eulerian_path(G))

# Plot the graph with the Eulerian trail
plt.figure(figsize=(6, 4))
pos = nx.circular_layout(G)  # Position nodes in a circle
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_edges(G, pos, edgelist=euler_trail, edge_color='red', width=2)
plt.title("Eulerian Trail")
plt.show()
```

An Eulerian trail is a path in a graph that traverses every edge exactly once but may not necessarily start and end at the same vertex. In simpler terms, it's a route through a graph that covers all the edges without repeating any, allowing you to start at one vertex and end at another.

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.cycle_graph(7)

# Add some extra edges to make it more complex
extra_edges = [(0, 3), (1, 4), (2, 5), (3, 6)]
G.add_edges_from(extra_edges)

# Find a Hamiltonian circuit
hamiltonian_circuit = list(G.nodes()) + [list(G.nodes())[0]]  # Hamiltonian circuit

# Plot the graph with the Hamiltonian circuit
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Position nodes with a spring layout
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_edges(G, pos, edgelist=[(hamiltonian_circuit[i], hamiltonian_circuit[i+1]) for i in range(len(hamiltonian_circuit)-1)], edge_color='red', width=2)
plt.title("Hamiltonian Circuit")
plt.show()
```

A Hamiltonian circuit is a path in a graph that visits each node exactly once and returns to the starting node, forming a closed loop. Think of it like a tour that covers all the stops in a city without repeating any, and eventually brings you back to where you started.

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
positions = {
    1: (0, 0),
    2: (1, 1),
    3: (2, 0),
    4: (1, -1),
    5: (3, 1),
    6: (4, 0),
}

for node, pos in positions.items():
    G.add_node(node, pos=pos)

edges = [
    (1, 2, 1.4),
    (1, 3, 2),
    (1, 4, 1.4),
    (2, 3, 1.4),
    (2, 4, 2),
    (2, 5, 2.2),
    (3, 4, 1.4),
    (3, 6, 2.2),
    (4, 6, 3),
    (5, 6, 1.4),
]

for edge in edges:
    G.add_edge(edge[0], edge[1], weight=edge[2])

pos = nx.get_node_attributes(G, 'pos')
nx.draw_networkx_nodes(G, pos, node_size=700)

edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edges(G, pos, edgelist=edges, width=2)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')

plt.title("Traveling Salesman Problem Graph")
plt.axis('off')
plt.show()
```

A Traveling Salesman Problem graph is a network of nodes connected by edges, where each edge has a weight representing the distance or cost of traveling between nodes. The goal is to find the shortest possible route that visits each node exactly once and returns to the starting point. The graph visually represents these nodes and weighted edges to help solve the problem.
